```python
import pandas as pd
```

```python
reviews = pd.read_csv("C:/Users/Ellis-pc/Documents/Projects/wine-reviews-exercise-Mellis21/data/winemag-data-130k-v2.csv")
```

```python
reviews.head()
```

```python
function_dict = {"description": "count", "points": "mean"}
```

```python
reviews_per_country = reviews.groupby(["country"]).aggregate(function_dict).sort_values(by = 'description', ascending = False)
```

```python
reviews_per_country
```

```python
reviews_per_country = pd.DataFrame(reviews_per_country)
```

```python
reviews_per_country = reviews_per_country.rename(columns = {"description":"count"})
```

```python
reviews_per_country
```

```python
reviews_per_country.points.round(1)
```

```python
reviews_per_country['points'] = reviews_per_country['points'].round(1)
```
